```python
import pandas as pd

# Read the Excel file
pharmacy_data = pd.read_excel('Pharmacy.xlsx')

def generate_insert_queries(dataframe, table_name):
    queries = []
    for index, row in dataframe.iterrows():
        item_code = row.get('Item_Code', '')
        description = row.get('Description', '')
        category = row.get('Category', '')
        cost_rate = row.get('CostRate', 0)
        sales_rate = row.get('SalesRate', 0)
        generic = row.get('Generic', '')

        # Generate SQL insert query
        query = f"INSERT INTO {table_name} (Item_Code, Description, Category, CostRate, SalesRate, Generic) VALUES ('{item_code}', '{description}', '{category}', {cost_rate}, {sales_rate}, '{generic}');"
        queries.append(query)
    return queries

# Generate SQL insert queries
insert_queries = generate_insert_queries(pharmacy_data, 'Product_TB')

# Print the SQL insert queries
for query in insert_queries:
    print(query)
    
    
    # Export SQL insert queries to a SQL file
# with open('insert_queries.sql', 'w') as sql_file:
    #for query in insert_queries:
       # sql_file.write(query + '\n')

# Export SQL insert queries to a text file
with open('insert_queries.txt', 'w') as text_file:
    for query in insert_queries:
        text_file.write(query + '\n')
```
